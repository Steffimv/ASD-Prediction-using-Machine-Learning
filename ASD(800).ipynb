```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import time
import warnings
warnings.filterwarnings('ignore')

# Load dataset
DATA_PATH = 'train.csv'
df = pd.read_csv(DATA_PATH)

# Optional: Rename target column if needed
if 'Class/ASD' in df.columns:
    df = df.rename(columns={'Class/ASD': 'target'})

# Replace '?' with NaN
df.replace('?', np.nan, inplace=True)

# Convert 'age' column to numeric and handle missing values
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['age'].fillna(df['age'].median(), inplace=True)

# Convert A1_Score to A10_Score to numeric and handle missing values
score_cols = [f'A{i}_Score' for i in range(1, 11)]
for col in score_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col].fillna(df[col].mode()[0], inplace=True)

# Encode categorical columns
le = LabelEncoder()
categorical_cols = ['gender', 'jaundice', 'austim', 'target']
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mode()[0])
        df[col] = le.fit_transform(df[col])

# Drop irrelevant columns if present
drop_cols = ['ethnicity', 'contry_of_res', 'used_app_before', 'age_desc', 'relation']
df = df.drop(columns=drop_cols, errors='ignore')

# Separate features and target
exclude_cols = ['target', 'result']  # 'result' is skipped if not present
X = df.drop(columns=exclude_cols, errors='ignore')
y = df['target']

# Check for any non-numeric columns
non_numeric_cols = X.select_dtypes(include=['object']).columns
if len(non_numeric_cols) > 0:
    raise ValueError(f"Non-numeric columns found: {non_numeric_cols}. All features must be numeric.")

# Define models
models = {
    'Logistic Regression': {'model': LogisticRegression(random_state=42, max_iter=1000), 'ml_technique': 'LR'},
    'SVM': {'model': SVC(random_state=42, probability=True), 'ml_technique': 'SVM'},
    'Decision Tree': {'model': DecisionTreeClassifier(random_state=42), 'ml_technique': 'DT'},
    'Random Forest': {'model': RandomForestClassifier(random_state=42), 'ml_technique': 'RF'},
    'XGBoost': {'model': XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss'), 'ml_technique': 'XGB'}
}

# Store results
results = []

# Evaluate models on both 70:30 and 80:20 splits
splits = {'70:30': 0.3, '80:20': 0.2}
for split_name, test_size in splits.items():
    print(f"\n=== {split_name} Split ===")
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    for model_name, config in models.items():
        model = config['model']
        ml_technique = config['ml_technique']
        print(f"\nResults for {model_name} (ML Technique: {ml_technique}):")

        # Train
        start_train = time.time()
        model.fit(X_train, y_train)
        end_train = time.time()

        # Test
        start_test = time.time()
        y_pred = model.predict(X_test)
        end_test = time.time()

        # Metrics
        accuracy = accuracy_score(y_test, y_pred)
        report = classification_report(y_test, y_pred, output_dict=True, zero_division=0)
        conf_matrix = confusion_matrix(y_test, y_pred)
        TN, FP, FN, TP = conf_matrix.ravel()
        specificity = TN / (TN + FP) if (TN + FP) > 0 else 0
        sensitivity = report['1']['recall']

        precision = f"{report['0']['precision']:.2f} (class 0), {report['1']['precision']:.2f} (class 1)"
        recall = f"{report['0']['recall']:.2f} (class 0), {report['1']['recall']:.2f} (class 1)"
        f1_score = f"{report['0']['f1-score']:.2f} (class 0), {report['1']['f1-score']:.2f} (class 1)"

        # ROC
        if hasattr(model, "predict_proba"):
            y_prob = model.predict_proba(X_test)[:, 1]
            fpr, tpr, _ = roc_curve(y_test, y_prob)
            roc_auc = auc(fpr, tpr)
        else:
            roc_auc = None

        # Important features
        if model_name == 'Logistic Regression':
            importance = pd.Series(model.coef_[0], index=X.columns).sort_values(ascending=False)
            important_features = ', '.join(importance.index)
        elif model_name in ['Random Forest', 'Decision Tree', 'XGBoost']:
            importance = pd.Series(model.feature_importances_, index=X.columns).sort_values(ascending=False)
            important_features = ', '.join(importance.index)
        else:
            important_features = 'N/A'

        # Display
        print(f"Accuracy: {accuracy * 100:.2f}%")
        print("Classification Report:", report)
        print(f"Specificity: {specificity:.2f}")
        print(f"Sensitivity: {sensitivity:.2f}")
        if roc_auc is not None:
            print(f"ROC AUC: {roc_auc:.2f}")
        print(f"Training Time: {end_train - start_train:.4f} seconds")
        print(f"Testing Time: {end_test - start_test:.4f} seconds")

        # Save results
        results.append({
            'ML Technique': ml_technique,
            'VALIDATION': split_name,
            'ACCURACY': f"{accuracy * 100:.2f}%",
            'PRECISION': precision,
            'RECALL': recall,
            'F1-SCORE': f1_score,
            'SENSITIVITY': f"{sensitivity:.2f}",
            'SPECIFICITY': f"{specificity:.2f}",
            'PLOT ROC CURVE': f"{roc_auc:.2f}" if roc_auc else 'N/A',
            'TRAINING TIME': f"{end_train - start_train:.4f} sec",
            'TESTING TIME': f"{end_test - start_test:.4f} sec",
            'Important Features': important_features
        })

        # Plot ROC
        if roc_auc is not None:
            plt.figure()
            plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
            plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
            plt.xlabel('False Positive Rate')
            plt.ylabel('True Positive Rate')
            plt.title(f'ROC Curve - {model_name} ({split_name})')
            plt.legend(loc='lower right')
            plt.grid()
            plt.show()

# Export results to Excel
results_df = pd.DataFrame(results)
results_df.to_excel('model_results(800).xlsx', index=False)
print("\nResults saved to 'model_results(800).xlsx'")
```
